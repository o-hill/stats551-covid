Replace all of the missing death values with 0 - we'll assume that an absence of reporting means that there were not any deaths for this project.

```python
for col in ['COVID-19 Deaths', 'Total Deaths', 'Pneumonia Deaths', 'Pneumonia and COVID-19 Deaths', 'Influenza Deaths', 'Pneumonia, Influenza, or COVID-19 Deaths']:
    df[col] = df[col].fillna(0)
```

```python
df.isnull().sum()
```

```python
df['Group'].unique()
```

```python
for group in df['Group'].unique():
    group_df = df[df['Group'] == group]
    group_df.to_csv(f'../../data/{group.split(" ")[-1].lower()}-deaths.csv')
```
